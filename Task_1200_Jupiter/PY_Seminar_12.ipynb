Дана задача:
К аналитику обратился риэлтор с задачей выставить ценник для продажи
объектов недвижимости площадью 48 кв. м. и 54 кв. м. В ходе работы
над задачей аналитик выяснил, что:
- объект 31 кв. м. стоит $19310
- объект 51 кв. м. стоит $52150
- объект 61 кв. м. стоит $74570
Вопрос: что сказать риэтору?
#20x**2 + 2x

```python
import matplotlib.pyplot as plt

x = [31, 51, 61]
y = [622.90, 1022.55, 1222.46]

fig, ax = plt.subplots()
ax.plot(x, y)
ax.grid()

plt.show()
```

```python
def f(x, price, meter, y):
    exp = ((meter) + (price) * (x - y)) * x
    print(round(exp, 2))
    
cost_per_meter_1 = 19310/31
cost_per_meter_2 = 52150/51

avarage_prices_1 = (52150/51 - 19310/31)/(51-31)
avarage_prices_2 = (74570/61 - 52150/51)/(61-51)
y_1 = 31
y_2 = 51

f(48, avarage_prices_1, cost_per_meter_1, y_1)
f(54, avarage_prices_2, cost_per_meter_2, y_2)
```

```python
from numpy import linalg as LA

# 31**2*a + 31*b + c = 19310
# 51**2*a + 31*b + c = 52150
# 61**2*a + 31*b + c = 74570

arr_a = [[31**2, 31, 1], [51**2, 51, 1], [61**2, 61, 1]]
arr_b = [19310, 52150, 74570]
x = LA.solve(arr_a, arr_b)
print(x)
```
